# # Classifiers comparison: decision trees and k-nearest neighbors on the dataset Iris


In the following program we compare the prediction results obtained by decision trees and k-nearest neighbors on the dataset Iris

The following cell shows the program training a decision tree and its results in preciction 

```python
from sklearn import tree 
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score # will be used to separate training and test
iris = load_iris()
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:1})
clf = clf.fit(iris.data, iris.target) # retaggio 
scores = cross_val_score(clf, iris.data, iris.target, cv=5) # score will be the accuracy
print(scores)
```

The following cell shows the training of k-nearest neighbors and its prediction results.
Here we use a uniform weighting setting (weights='uniform'): any neighbors weights the same in the majority voting aggregation.

```python
from sklearn import neighbors
n_neighbors = 11
clf_knn = neighbors.KNeighborsClassifier(n_neighbors, weights='uniform')
clf_knn = clf_knn.fit(iris.data, iris.target)
scores = cross_val_score(clf_knn, iris.data, iris.target, cv=5) # score will be the accuracy
print(scores)
# shows the model predictions  
for i in range(len(iris.target)):
    print(iris.data[i,:])
    print(iris.data[i,:].reshape(1,-1))
    instance=iris.data[i,:].reshape(1,-1)
    #print(clf_knn.predict(instance))
    predicted=clf_knn.predict(instance)[0]
    print(predicted)
    if iris.target[i]==predicted:
        print(str(i)+" ok "+str(iris.target_names[iris.target[i]]))
    else:
        print(str(i)+" nok "+"true class: "+str(iris.target_names[iris.target[i]])+"; predicted: "+str(iris.target_names[predicted]))
```

The following cell shows the tuning of the k-nn models with a varying value of k (number of nearest neighbors) and finds the best value of k (giving the maximum accuracy).

```python
import numpy as np
from numpy import zeros
from sklearn import neighbors
from sklearn.datasets import load_iris
iris = load_iris()
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0)

best_accuracy=0
best_k=1
A=np.zeros(len(y_train), dtype=float) # for storing accuracies
for n_neighbors in np.arange(1,len(y_train)+1):
    clf_knn3 = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
    # (n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=1, **kwargs)
    clf_knn3.fit(X_train, y_train)
    index=n_neighbors-1
    A[index]=clf_knn3.score(X_test, y_test)
    if best_accuracy<clf_knn3.score(X_test, y_test):
        best_accuracy=clf_knn3.score(X_test, y_test)
        best_k=n_neighbors
    print("k neighbors="+str(n_neighbors))
    print("accuracy="+str(clf_knn3.score(X_test, y_test)))
    
print("\n")
print("best k="+str(best_k))
print("best accuracy="+str(best_accuracy))

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#plt.xticks(np.arange(1, len(y_train)+1, 8))
plt.yticks(np.arange(0.0,1.0,0.01))

plt.plot(np.arange(1,len(y_train)+1),A)

plt.show()
```

1. Plot the Iris dataset, in 2-D, with the color as above determined by the k-nn estimation

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import neighbors
from sklearn.model_selection import cross_val_predict
from itertools import combinations
from sklearn.model_selection import cross_val_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Create a k-NN classifier
n_neighbors = 11
clf_knn = neighbors.KNeighborsClassifier(n_neighbors, weights='uniform')

# Get predicted labels using cross-validation
predicted = cross_val_predict(clf_knn, X, y, cv=5)

# Convert class labels to colors
colors = np.array(['red', 'blue', 'green'])[predicted]

feature_combinations = list(combinations(range(X.shape[1]), 2))

# Calculate and print accuracy for each feature combination
for feature1, feature2 in feature_combinations:
    X_selected = X[:, [feature1, feature2]]
    accuracy_scores = cross_val_score(clf_knn, X_selected, y, cv=5)
    avg_accuracy = np.mean(accuracy_scores)
    print(f"Features: {iris.feature_names[feature1]}, {iris.feature_names[feature2]} - Average Accuracy: {avg_accuracy:.2f}")

# Create subplots
plt.figure(figsize=(15, 10))
plot_num = 1
for feature1 in range(X.shape[1]):
    for feature2 in range(feature1 + 1, X.shape[1]):
        plt.subplot(2, 3, plot_num)  # 2 rows, 3 columns of subplots

        # Scatter plot with colors based on iris species
        plt.scatter(X[:, feature1], X[:, feature2], c=colors)
        plt.xlabel(iris.feature_names[feature1])
        plt.ylabel(iris.feature_names[feature2])
        plt.title(f"{iris.feature_names[feature1]} vs {iris.feature_names[feature2]}")

        plot_num += 1

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from Utils import find_best_k



# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Binarize the labels for each class
y_bin = label_binarize(y, classes=[0, 1, 2])
n_classes = y_bin.shape[1]

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.4, random_state=0)


# Choose the best value of k (you need to determine this)
best_k = find_best_k(X_train,y_train)

# Train a k-NN classifier for each class (one-vs-all)
classifiers = [KNeighborsClassifier(n_neighbors=best_k) for _ in range(n_classes)]
for i in range(n_classes):
    classifiers[i].fit(X_train, y_train[:, i])

# Compute ROC curves and ROC AUC for each class (one-vs-all)
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    y_scores = classifiers[i].predict_proba(X_test)[:, 1]  # Positive class probability
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_scores)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves for each class
plt.figure(figsize=(10, 7))
colors = ['blue', 'red', 'green']
for i, color in zip(range(n_classes), colors):
    class_name = iris.target_names[i]  # Get the class name
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'{class_name} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f"ROC One-vs-Rest con best k = {best_k}")
plt.legend(loc="lower right")
plt.show()



```

4. Compare decision trees and k-nn on the ROC space: for which values of (TPR,FPR) k-nn is better than decision trees?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc

# Carica il dataset Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Dividi il dataset in set di allenamento e set di test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crea il modello Decision Tree
tree_classifier = DecisionTreeClassifier(class_weight='balanced', criterion='entropy', max_features='sqrt', max_leaf_nodes=5, min_impurity_decrease=0, min_samples_leaf=1, min_samples_split=5, splitter='random')

# Valutazione con cross-validation per il Decision Tree
tree_y_prob = cross_val_predict(tree_classifier, X_train, y_train, cv=5, method='predict_proba')

# Crea il modello K-Nearest Neighbors
knn_classifier = KNeighborsClassifier(n_neighbors=13)

# Valutazione con cross-validation per il K-Nearest Neighbors
knn_y_prob = cross_val_predict(knn_classifier, X_train, y_train, cv=5, method='predict_proba')

# Calcola le curve ROC per entrambi i modelli e le AUC
n_classes = 3
tree_fpr = dict()
tree_tpr = dict()
tree_auc = dict()
knn_fpr = dict()
knn_tpr = dict()
knn_auc = dict()

for i in range(n_classes):
    # Decision Tree
    tree_fpr[i], tree_tpr[i], _ = roc_curve(y_train == i, tree_y_prob[:, i])
    tree_auc[i] = auc(tree_fpr[i], tree_tpr[i])

    # K-Nearest Neighbors
    knn_fpr[i], knn_tpr[i], _ = roc_curve(y_train == i, knn_y_prob[:, i])
    knn_auc[i] = auc(knn_fpr[i], knn_tpr[i])

# Plotting delle curve ROC per ciascuna classe
plt.figure(figsize=(8, 6))
colors = ['darkorange', 'navy', 'cornflowerblue']

for i, color in zip(range(n_classes), colors):
    plt.plot(tree_fpr[i], tree_tpr[i], color=color, lw=2, label=f'Decision Tree Class {i} (AUC = {tree_auc[i]:.2f})')
    plt.plot(knn_fpr[i], knn_tpr[i], color=color, lw=2, linestyle='--', label=f'KNN Class {i} (AUC = {knn_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Multiclass with Cross-Validation')
plt.legend(loc='lower right')
plt.show()
```
